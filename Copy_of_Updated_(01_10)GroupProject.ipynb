#Filtered out unused columns from original df

```python
fifa = df[['ID','Name','Age','Nationality','Overall','Club','Wage','Preferred Foot','Height','Weight']]
```

```python
fifa
```

#Rename a column

```python
fifa.rename(columns={"ID":"Player ID"}, inplace=True) 
```

#Rename a column again

```python
fifa.rename(columns={"Height":"Height in Feet"}, inplace=True) 
```

```python
fifa
```

#Create a new column
######(Reformated values of 'Weight' to remove 'lbs')

```python
fifa['Weight in Kg']=fifa['Weight'].str.strip("lbs")  
```

```python
fifa
```

#Broadcasted to update newly added column

```python
#converted column to a number to be able to broadcast 
fifa['Weight in Kg'] = pd.to_numeric(fifa['Weight in Kg']).mul(0.453592).round(2)   
fifa
```

```python
#shows change from object to float64
fifa.dtypes
```

#Made copy of DataFrame

```python
fifa2=fifa.copy()
```

```python
#removing unnecessary characters
fifa2['Wage']=fifa2['Wage'].str.strip("K").str.strip("€")
```

```python
#change type from object to float64 and added new column
fifa2['Wage in Thousands of Pounds'] = pd.to_numeric(fifa2['Wage'])
```

```python
#delete entire 'wage' col
fifa2.drop(['Wage'],axis=1, inplace=True)
```

```python
fifa2
```

#1: highest paid players in soccer (top 5)

```python
fifa2.set_index('Name').nlargest(5,'Wage in Thousands of Pounds')
```

#2: Most efficient player in the sport

```python
fifa2.set_index('Name').nlargest(1, 'Overall')
```

# 3: Better team in soccer

```python
#team club with the highest average overall, since multiple players from same team...
    #fifa2['Club'].value_counts()
best=fifa2[['Club','Overall']].groupby('Club').mean()
best.nlargest(1,'Overall')
```

#Q1: Top 5 highest paid teams

```python
teams=fifa2[['Wage in Thousands of Pounds']].groupby(fifa2['Club']).sum()
teams.nlargest(5,'Wage in Thousands of Pounds')
```

#Q2: Top 5 Overall Teams

```python
overall= fifa2[['Overall']].groupby(fifa2['Club']).mean()
overall.nlargest(5, 'Overall')
```
