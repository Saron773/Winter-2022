```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/redun1/ML_Data/main/data.csv')
```

```python
#added a column at the end using the data from id
df['Player ID'] = df['ID']
```

```python
df.head()
```

```python
df.rename(columns={"ID":"Player ID"}) #changes col name
```

```python
#isolated the columsn we wanted from the original dataset
fifa = df[['ID','Name','Age','Nationality','Overall','Club','Wage','Preferred Foot','Height','Weight']]
```

```python
fifa
```

```python
fifa.rename(columns={"ID":"Player ID"}, inplace=True) #rename a col 
```

```python
fifa.rename(columns={"Height":"Height in Feet"}, inplace=True) #rename a col
```

```python
fifa
```

```python
fifa['Weight in Kg']=fifa['Weight'].str.strip("lbs")  #remove unnec. label at end of col

#fifa['Weight'].str.strip("lbs")
#int(fifa['Weight'].str.strip("lbs"))
```

```python
fifa
```

```python
fifa['Weight in Kg'] = pd.to_numeric(fifa['Weight in Kg']).mul(0.453592).round(2)  #convert col to a number to be able to broadcast 
```

```python
fifa
```

```python
#made a copy of the dataframe
fifa2=fifa.copy()
```

```python
fifa2['Wage']=fifa2['Wage'].str.strip("K")
```

```python
fifa2['Wage']=fifa2['Wage'].str.strip("€")
```

```python
fifa2
```

```python
fifa2['Wage in Thousands of Pounds'] = pd.to_numeric(fifa2['Wage'])
```

```python
fifa2.drop(['Wage'],axis=1, inplace=True)
```

```python
fifa2
```

```python
#the 5 players with the highest salaries are:

#fifa2[['Wage in Thousands of Pounds']].sort()

fifa2.nlargest(5,['Wage in Thousands of Pounds'])
```
