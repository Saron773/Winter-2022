```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
apartments = pd.read_csv('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/property.csv')
```

```python
#apartments.loc[4].fillna(0, axis = 0, inplace = True)
```

```python
apartments['NUM_BEDROOMS'].fillna(0, inplace=True) #fill zero for NaN values in bedrooms
```

```python
apartments
```

```python
apartments['NUM_BATH'].fillna(1, inplace=True)  #fill one for NaN values in bathrooms
```

```python
apartments
```

```python
apartments2 = apartments.replace('HURLEY',1) #replace HURLEY in NUM_BATH with one
apartments2
```

```python
apartments2['OWN_OCCUPIED'].replace('12','Unknown',inplace=True) #replace 12 with unknown
```

```python
apartments2['OWN_OCCUPIED'].fillna('Unknown',inplace = True) #fillin NAN with unknown 
```

```python
apartments2
```

```python
apartments2.groupby(['ST_NAME'])['SQ_FT'].count().sort_values(ascending=True).plot(kind='bar')
```

```python
apartments2.groupby(['ST_NAME'])['NUM_BEDROOMS'].count().sort_values(ascending=True).plot(kind='barh')
```

```python
apartments2.groupby(['ST_NAME'])['NUM_BATH'].count().sort_values(ascending=True).plot(kind='barh')
```
